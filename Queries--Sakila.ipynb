# Queries

Queries 1

```python
q = """
select first_name, last_name, email, adr.address, city, cty.city_id
from customer as cus
join address as adr
on cus.address_id = adr.address_id
join city as cty 
on adr.city_id = cty.city_id
where cty.city_id = 312
"""
pd.read_sql(q,engine)
```

Queries 2

```python
q = """
select fi.title, fi.description, fi.release_year, fi.rating, fi.special_features
from category as cat
join film_category as fil_cat
on cat.category_id = fil_cat.category_id
join film as fi
on fil_cat.film_id = fi.film_id
"""
pd.read_sql(q,engine)
```

Queries 3

```python
q = """
select ac.last_name, fi.title, fi.release_year
from actor as ac
join film_actor as fi_act
on ac.actor_id = fi_act.actor_id
join film as fi
on fi_act.film_id = fi.film_id
where ac.last_name = 'LOLLOBRIGIDA'
"""
pd.read_sql(q,engine)
```

Queries 4

```python
q = """
select ac.first_name, ac.last_name, fi.title, fi.release_year
from actor as ac
join film_actor as fi_act
on ac.actor_id = fi_act.actor_id
join film as fi
on fi_act.film_id = fi.film_id
where fi.title = 'Bingo Talented'
"""
pd.read_sql(q,engine)
```

Queries 5

```python
q = """
select cus.customer_id, pay.amount
from customer as cus
join payment as pay
on cus.customer_id = pay.customer_id
where pay.amount < "2*"
"""
pd.read_sql(q,engine)
```

Queries 6

```python
q = """
select
    customer.first_name,
    customer.last_name,
    (select
            count(*)
        from
            payment
        where
            payment.customer_id = customer.customer_id) as num_payments
from
    customer
order by num_payments DESC
limit 5;
"""
    
pd.read_sql(q,engine)
```
